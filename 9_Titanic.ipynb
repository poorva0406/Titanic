```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('~/Downloads/train.csv')
df.head(10)
```

```python
fig, ax = plt.subplots(figsize=(20, 13))
box = sns.boxplot(x='Sex',y='Age',hue='Survived',data=df,palette='Set2',showfliers=True)
ax.yaxis.grid(True)
plt.show()
```

#### Male around the age of 18-36 have a higher survival rate than other males
#### Female around the age of 19-38 have a higher survival rate than other females
